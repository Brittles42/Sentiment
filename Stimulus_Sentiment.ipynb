```python
#labelling the sentiments within text as positive or negative
from transformers import pipeline
import pandas as pd
import seaborn 
```

```python
pipe = pipeline("text-classification", model="distilbert-base-uncased-finetuned-sst-2-english")
text_data = [
    "I recently attended a business conference, and it was an insightful event. The speakers shared valuable industry insights, and I made valuable connections.",
    "My meeting with a potential client went exceptionally well. We had a productive discussion, and they expressed strong interest in our services.",
    "I had the opportunity to collaborate with a talented team on a project, and the synergy was remarkable. Our combined efforts led to a successful outcome.",
    "Our company's quarterly results were impressive, exceeding our projections. It's a testament to our team's hard work and dedication.",
    "I participated in a negotiation that resulted in a beneficial partnership for our company. It was a win-win situation for both parties involved.",
    "The training program I attended was well-structured and informative. I gained new skills that I can apply to my role effectively.",
    "I received positive feedback from a client about our customer support. It's gratifying to know that our team is providing exceptional service.",
    "Our recent product launch was a hit in the market. Customers are raving about its features and functionality.",
    "I had a productive brainstorming session with colleagues, and we came up with innovative ideas to improve our processes.",
    "Our company's commitment to sustainability and corporate social responsibility is commendable. It's rewarding to be part of an environmentally conscious and inclusive organization.",
    "I had the privilege of mentoring a junior team member, and I'm proud of their growth and contributions to the team.",
    "Our company's culture of collaboration and teamwork fosters a positive work environment. It's one of the reasons I enjoy coming to work every day.",
    "I received recognition for my contributions to a project during our team meeting. It's motivating to know that hard work doesn't go unnoticed.",
    "I had a chance to attend a networking event, and I connected with professionals who have valuable insights to share.",
    "Our company's commitment to employee development is evident. I appreciate the opportunities for learning and growth."
    "I had a frustrating experience.",
    "Our recent marketing campaign didn't yield the expected results, and we faced disappointing sales numbers.",
    "A project I was leading faced unexpected delays, causing frustration among team members and stakeholders.",
    "I encountered technical issues during an important video conference, which disrupted the meeting.",
    "Our company faced a budget cut, leading to layoffs and increased workload for remaining employees.",
    "I received negative feedback from a supervisor regarding a project I worked on, and it was disheartening.",
    "A key client decided to terminate their contract with our company, leading to a loss in revenue.",
]
#applying the model and saving the output into a new dataframe.
output_data = pipe(text_data)
experience_df = pd.DataFrame(output_data)
experience_df.label.value_counts()
```

```python
#graphing results
experience_df['label'].value_counts().plot(kind='bar').set(title = 'Frequency of Experiences', 
                                                       xlabel = 'Experience', ylabel = 'Count')
```

```python
#visualizing distribution of scores   
seaborn.boxplot(x = 'score', data = experience_df).set(title = 'Accuracy Distribution')
```
